# IBB Hava kalitesi verisini cekme

```python
import requests

url='https://cevrekoruma.ibb.gov.tr/HavaKaliteInceleme/Havakaliterapor.aspx'

# in case you need a session
#cd = { 'sessionid': '123..'}

#r = requests.get(url, cookies=cd)
r = requests.get(url)
# or without a session: r = requests.get(url)
a= str(r.content)
```

## Verileri Istasyonlara gore parcalama 

```python
basla = a.find("[[")
sonlanma = a.find("]]")

user_input = a[basla:sonlanma]

parcalar = user_input.split("],[")
```

## Istasyon verilerini cekme
